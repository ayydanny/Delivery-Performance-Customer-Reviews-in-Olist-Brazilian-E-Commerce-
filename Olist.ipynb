## Main Question: What determines low vs high customer ratings in Olist e-commerce?

#### Key Explorations:
1. Sellers
2. Geolocation

#### Supporting variable
- Delivery time

```python
sellers_query = pd.read_sql("""
WITH CTE AS (
    SELECT
        c.customer_id,
        s.seller_id,
        o.order_status,
        o.order_purchase_timestamp,
        o.order_approved_at,
        o.order_delivered_customer_date,
        o.order_estimated_delivery_date,
        ore.review_score
    FROM customers c
    JOIN orders o
        ON o.customer_id = c.customer_id
    JOIN order_items oi
        ON oi.order_id = o.order_id
    JOIN sellers s
        ON s.seller_id = oi.seller_id
    JOIN order_reviews ore
        ON ore.order_id = oi.order_id
)
SELECT
    *,
    (julianday(order_approved_at) - julianday(order_purchase_timestamp)) * 24
        AS approval_time_hours,
    julianday(order_delivered_customer_date) - julianday(order_estimated_delivery_date) AS delivery_delay_days
FROM CTE
""", conn)
```

```python
sellers_query.to_sql("sellers_query", conn, if_exists="replace", index=False)
```

```python
sellers_final = pd.read_sql("""
SELECT
    seller_id,
    COUNT(*) as total_orders,
    AVG(approval_time_hours) AS AVG_approval_time_hours,
    AVG(delivery_delay_days) AS AVG_delivery_delay_days,
    AVG(review_score) AS AVG_review_score
FROM sellers_query
WHERE order_status = 'delivered'
GROUP BY seller_id
HAVING total_orders > 5
ORDER BY total_orders DESC
""", conn)

sellers_final.to_csv("sellers_final.csv", index=False)
```

```python
sellers_final
```

```python
geolocation_query = pd.read_sql("""
WITH CTE AS (
    SELECT
        c.customer_state,
        o.order_status,
        o.order_purchase_timestamp,
        o.order_approved_at,
        o.order_delivered_customer_date,
        o.order_estimated_delivery_date,
        ore.review_score
    FROM customers c
    JOIN orders o
        ON o.customer_id = c.customer_id
    JOIN order_reviews ore
        ON ore.order_id = o.order_id
)
SELECT
    *,
    (julianday(order_approved_at) - julianday(order_purchase_timestamp)) * 24
        AS approval_time_hours,
    julianday(order_delivered_customer_date) - julianday(order_estimated_delivery_date)
        AS delivery_delay_days
FROM CTE
""", conn)
```

```python
geolocation_query.to_sql("geolocation_query", conn, if_exists="replace", index=False)
```

```python
geolocation_final = pd.read_sql("""
SELECT
    COUNT(*) AS total_orders,
    customer_state,
    AVG(approval_time_hours) AS AVG_approval_time_hours,
    AVG(delivery_delay_days) AS AVG_delivery_delay_days,
    AVG(review_score) AS AVG_review_score
FROM geolocation_query
WHERE order_status = 'delivered'
GROUP BY customer_state
""", conn)

geolocation_final.to_csv("geolocation_final.csv", index=False)
```

```python
sellers_final.to_sql("sellers_final", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("""
SELECT
    AVG(AVG_review_score) AS avg_review_all_sellers
FROM sellers_final
""", conn)
```

```python
pd.read_sql("""
WITH CTE AS (
    SELECT
        *,
        AVG(AVG_approval_time_hours) OVER () AS avg_review_approval_sellers,
        AVG(AVG_delivery_delay_days) OVER () AS avg_review_delivery_sellers
    FROM sellers_final
)
SELECT
    AVG(AVG_review_score) AS avg_review_fast_approval_sellers
    FROM cte
    WHERE AVG_approval_time_hours < avg_review_approval_sellers
""", conn)
```

```python
pd.read_sql("""
WITH CTE AS (
    SELECT
        *,
        AVG(AVG_approval_time_hours) OVER () AS avg_review_approval_sellers,
        AVG(AVG_delivery_delay_days) OVER () AS avg_review_delivery_sellers
    FROM sellers_final
)
SELECT
    AVG(AVG_review_score) AS avg_review_fast_delivery_sellers
    FROM cte
    WHERE AVG_delivery_delay_days < avg_review_delivery_sellers
""", conn)
```

```python
pd.read_sql("""
WITH CTE AS (
    SELECT
        *,
        AVG(AVG_approval_time_hours) OVER () AS avg_review_fast_approval_sellers,
        AVG(AVG_delivery_delay_days) OVER () AS avg_review_fast_delivery_sellers
    FROM sellers_final
)
SELECT
    AVG(AVG_review_score) AS avg_review_fast_approval_and_delivery
    FROM cte
    WHERE AVG_delivery_delay_days < avg_review_fast_delivery_sellers AND AVG_approval_time_hours < avg_review_fast_approval_sellers
""", conn)
```

The average review score across all sellers is approximately 4.14.

Sellers with faster-than-average approval times have a slightly higher average review score (≈ +0.02).

Sellers with faster-than-average delivery times show a more noticeable increase in review scores (≈ +0.06).

When sellers are faster than average in both approval and delivery, the average review score increases the most (≈ +0.07).

Conclusion:
While review scores are subjective and influenced by factors outside a seller’s control, faster delivery times appear to have a stronger relationship with higher customer reviews than approval speed. This is likely because approval times vary by hours, whereas delivery delays vary by multiple days, making delivery more salient to customers.
